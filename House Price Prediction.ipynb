# Preprocessing function

```python
#Creat preprocessing function to select columns and deal with missing values
def preprocessing(df): 
    #Fillna with MasVnrArea
    df['MasVnrArea'].fillna('0.0', inplace=True)
    #Fillna with BsmtQual
    B1=df[(df['BsmtQual'].isna()==True) & (df['YearBuilt']>=1990)]
    B1['BsmtQual'].fillna('GA', inplace=True)
    df['BsmtQual'].fillna('TA', inplace=True)
    #Fillna with GarageType
    df['GarageType'].fillna('Attchd', inplace=True)
    #Fillna with GarageYrBlt
    df['GarageYrBlt'].fillna(df['YearBuilt'], inplace=True)
    #Select columns 
    df=df[['OverallQual','YearBuilt','YearRemodAdd','MasVnrArea','TotalBsmtSF','1stFlrSF','GrLivArea','FullBath',
            'TotRmsAbvGrd','Fireplaces','GarageYrBlt','GarageCars','GarageArea','MSZoning','Neighborhood','BldgType',
            'HouseStyle','RoofMatl','ExterQual','BsmtQual','Heating','CentralAir','KitchenQual','GarageType','SaleCondition']]
    df['GarageYrBlt']=df['GarageYrBlt'].astype('int')
    return df
```

```python
#Preprocessing train data
selected_df=preprocessing(train)
```

# Modeling

```python
#Import basic modeling package
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

#### I separate the preprocessed train data to numerical part and categorical part. For the numerical part I transform them to be standardized.

```python
#Standardize selected Nunerical columns
x_num=selected_df[['OverallQual','YearBuilt','YearRemodAdd','MasVnrArea','TotalBsmtSF','1stFlrSF','GrLivArea','FullBath',
            'TotRmsAbvGrd','Fireplaces','GarageYrBlt','GarageCars','GarageArea']]
SCALE=StandardScaler()
SCALE.fit(x_num)
x_num_scale=SCALE.transform(x_num)
#Rebulid nparray to dataframe
column_name=['OverallQual','YearBuilt','YearRemodAdd','MasVnrArea','TotalBsmtSF','1stFlrSF','GrLivArea','FullBath',
            'TotRmsAbvGrd','Fireplaces','GarageYrBlt','GarageCars','GarageArea']
x_num_scale=pd.DataFrame(x_num_scale,columns=column_name)
x_num_scale
```

#### For the categorical part, I use LabelEncoder to transform the value to number label. Then I combine the numerical part and categorical part, and execute train test split.  

```python
#Preprocess selected categorical columns
x_cat=selected_df[['MSZoning','Neighborhood','BldgType',
            'HouseStyle','RoofMatl','ExterQual','BsmtQual','Heating','CentralAir','KitchenQual','GarageType','SaleCondition']]
#label encode categorical columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_cat_dum = x_cat.apply(le.fit_transform)
#Combine numerical and categorical columns together
X= pd.concat([x_num_scale, x_cat_dum], axis=1)
#Define target
y=train[['SalePrice']]
#Train test split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

#### Then I try different types ML model.

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
yhat=lm.predict(x_test)
lm.score(x_test,y_test)
```

## Polynomial Rgression 

```python
from sklearn.preprocessing import PolynomialFeatures
order=[1,2,3,4,5]
score=[]
for n in order:
    lm=LinearRegression()
    pr=PolynomialFeatures(degree=n,include_bias=False)
    x_train_pr=pr.fit_transform(x_train)
    x_test_pr=pr.fit_transform(x_test)
    lm.fit(x_train_pr,y_train)
    lm.predict(x_test_pr)
    score.append(lm.score(x_test_pr,y_test))
score
```

## Ridge Regression 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
param=[{'alpha':[0.01, 0.1,1,10,15,50,100,200]}]
r=Ridge(normalize=True)
gs=GridSearchCV(r,param,cv=4)
gs.fit(x_train, y_train)
yhat=gs.predict(x_test)
gs.best_estimator_
```

```python
r=Ridge(alpha=0.1,normalize=True)
r.fit(x_train, y_train)
yhat=r.predict(x_test)
r.score(x_test,y_test)
```

## Ridge Regression with Polynomial Features 

```python
alpha_space=[0.01,0.1,1,10,15,50,100,200]
param_grid={'alpha':alpha_space}
order=[1,2,3]
estimator=[]
for n in order:
    pr=PolynomialFeatures(degree=n,include_bias=False)
    x_train_pr=pr.fit_transform(x_train)
    x_test_pr=pr.fit_transform(x_test)
    r=Ridge(normalize=True)
    lcv=GridSearchCV(r,param_grid,cv=5)
    lcv.fit(x_train_pr,y_train)
    lcv.predict(x_test_pr)
    bes=lcv.best_estimator_
    degree=n
    scores=lcv.score(x_test_pr,y_test)
    estimator.append(str(degree))
    estimator.append(bes)
    estimator.append(scores)
estimator
```

```python
pr=PolynomialFeatures(degree=3,include_bias=False)
x_polly=pr.fit_transform(x_train)
x_test_polly=pr.fit_transform(x_test)
r=Ridge(alpha=15,normalize=True)
r.fit(x_polly, y_train)
yhat=r.predict(x_test_polly)
r.score(x_test_polly,y_test)
```

### Update prerocessing function

```python
#Update preprocessing function 
def preprocessing(df): 
    #Fillna with MasVnrArea
    df['MasVnrArea'].fillna('0.0', inplace=True)
    #Fillna with BsmtQual
    B1=df[(df['BsmtQual'].isna()==True) & (df['YearBuilt']>=1990)]
    B1['BsmtQual'].fillna('GA', inplace=True)
    df['BsmtQual'].fillna('TA', inplace=True)
    #Fillna with GarageType
    df['GarageType'].fillna('Attchd', inplace=True)
    #Fillna with GarageYrBlt
    df['GarageYrBlt'].fillna(df['YearBuilt'], inplace=True)
    #Fillna with GarageCars
    df['GarageCars'].fillna(2, inplace=True)
    #Fillna with GarageArea
    df['GarageArea'].fillna(480, inplace=True)
    #Fillna with MSZoning
    df['MSZoning'].fillna('RL', inplace=True)
    #Fillna with TotalBsmtSF
    df['TotalBsmtSF'].fillna(df['1stFlrSF'], inplace=True)
    #Fillna with KitchenQual
    K1=df[df['Utilities']=='NoSeWa']
    K1['KitchenQual'].fillna('Gd', inplace=True)
    df['KitchenQual'].fillna('TA', inplace=True)
    #Select columns 
    df=df[['OverallQual','YearBuilt','YearRemodAdd','MasVnrArea','TotalBsmtSF','1stFlrSF','GrLivArea','FullBath',
            'TotRmsAbvGrd','Fireplaces','GarageYrBlt','GarageCars','GarageArea','MSZoning','Neighborhood','BldgType',
            'HouseStyle','RoofMatl','ExterQual','BsmtQual','Heating','CentralAir','KitchenQual','GarageType','SaleCondition']]
    df['GarageYrBlt']=df['GarageYrBlt'].astype('int')
    return df
```

```python
selected_df=preprocessing(test)
x_num=selected_df[['OverallQual','YearBuilt','YearRemodAdd','MasVnrArea','TotalBsmtSF','1stFlrSF','GrLivArea','FullBath',
            'TotRmsAbvGrd','Fireplaces','GarageYrBlt','GarageCars','GarageArea']]
SCALE=StandardScaler()
SCALE.fit(x_num)
x_num_scale=SCALE.transform(x_num)
#Rebulid nparray to dataframe
column_name=['OverallQual','YearBuilt','YearRemodAdd','MasVnrArea','TotalBsmtSF','1stFlrSF','GrLivArea','FullBath',
            'TotRmsAbvGrd','Fireplaces','GarageYrBlt','GarageCars','GarageArea']
x_num_scale=pd.DataFrame(x_num_scale,columns=column_name)
#Preprocess selected categorical columns
x_cat=selected_df[['MSZoning','Neighborhood','BldgType',
            'HouseStyle','RoofMatl','ExterQual','BsmtQual','Heating','CentralAir','KitchenQual','GarageType','SaleCondition']]
le = LabelEncoder()
x_cat_dum = x_cat.apply(le.fit_transform)
#Combine numerical and categorical columns together
X= pd.concat([x_num_scale, x_cat_dum], axis=1)
```
